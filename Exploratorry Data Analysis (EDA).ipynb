## Menganalisis Variabel Secara Individual

```python
numerical_vars = ['no_of_trainings', 'age', 'length_of_service', 'avg_training_score']
categorical_vars = ['department', 'region', 'education', 'recruitment_channel', 'performance_rating', 'KPIs_met >80%', 'awards_won?', 'is_promoted']
```

```python
data = data[numerical_vars+categorical_vars]
```

```python
data.shape
```

## Variabel Numerikal

```python
data[numerical_vars].describe()
```

```python
data[numerical_vars].hist(edgecolor='black', bins=30, figsize=(10, 5), layout = (2,2));
```

## Variabel Kategorikal

```python
data[categorical_vars].describe()
```

```python
fig, ax = plt.subplots(2,4, figsize=(14,6))
for var, subplot in zip(categorical_vars, ax.flatten()):
    data[var].value_counts().plot(kind='bar', ax=subplot, title=var)
    
    fig.tight_layout()
```

## Hubungan Antara Variabel dengan Scatter Plot

```python
data.plot.scatter(x='avg_training_score', y='length_of_service');
```

```python
sns.jointplot(x='avg_training_score', y='length_of_service', data=data, joint_kws={"s": 10});
```

```python
sns.pairplot(data[numerical_vars[:4]], plot_kws={"s": 10});
```

```python
data[numerical_vars].corr()
```

```python
data[numerical_vars].corr()['length_of_service'].sort_values(ascending=False)
```

```python
correlations = data[numerical_vars].corr()
```

```python
fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(correlations, ax = ax);
```

## Hubungan Antara Variabel dengan Box Plot

```python
fig, ax = plt.subplots(3,3, figsize=(14,9))
for var, subplot in zip(categorical_vars, ax.flatten()):
    sns.boxplot(x='performance_rating', y='length_of_service', data=data, ax=subplot)
    
    fig.tight_layout()
```

```python
fig, ax = plt.subplots(3,3, figsize=(14,9))
for var, subplot in zip(categorical_vars, ax.flatten()):
    sns.boxplot(x='department', y='length_of_service', data=data, ax=subplot)
    
    fig.tight_layout()
```
